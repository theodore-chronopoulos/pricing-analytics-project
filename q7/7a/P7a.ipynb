## 7a. AI-generated booking column

I used ChatGPT (GPT-5.4 Thinking) as the AI agent. Because prompting all 8,354 search queries was not feasible in-chat, I implemented the AI choice-generation step on a random sample of 100 search queries using random seed 42. For each sampled `srch_id`, I retained all hotel alternatives in the customer’s choice set and asked the AI to choose either one listed hotel `alt_id` or `NO_PURCHASE`.

The AI choice was converted into a binary booking column as follows. If the AI selected a hotel alternative, that row received `booking_ai_sample100_corrected = 1` and all other alternatives in the same search query received 0. If the AI selected `NO_PURCHASE`, all hotel alternatives in that search query received 0.

The corrected AI-generated sample contains 100 search queries and 2,036 hotel rows. The AI selected a hotel in 80 queries and selected `NO_PURCHASE` in 20 queries.

### Prompt Explanation:
The prompt instructed the AI agent to predict hotel booking behavior using the customer/search context and hotel attributes. For each query, the AI was shown the booking window, number of adults, number of children, number of rooms, Saturday-night indicator, and the full set of hotel alternatives with star rating, review score, brand indicator, location score, accessibility score, historical price, displayed price, and promotion flag. The AI was asked to return exactly one listed `alt_id` or `NO_PURCHASE`.

### Subset Explanation:
Because the AI-generated dataset was created for a 100-query sample rather than the full dataset, the results should be interpreted as a sample-based implementation of the assignment rather than a full AI relabeling of all 8,354 search queries.
