## 7d. Discussion

The results suggest that AI agents can capture some broad patterns in hotel choice behavior, but they are not a perfect substitute for models estimated from real human choices. In the AI-generated dataset, the choices implied intuitive preferences: the AI tended to favor hotels with higher star ratings, better review scores, better location scores, brand status, accessibility, and promotions, while penalizing higher displayed prices and higher historical prices. These signs are broadly consistent with the human MNL estimates, suggesting that the AI agent understands the basic economic tradeoffs in the hotel search setting.

However, the AI-generated choices appeared more deterministic and more strongly attribute-driven than real human choices. When the MNL was re-estimated on the AI-generated booking column, the AI-implied coefficients were generally larger in magnitude than the human-data coefficients. This suggests that the AI relied heavily on observable features such as price, review score, location, and star rating, whereas real customers likely respond to unobserved factors such as hotel photos, loyalty programs, brand familiarity, map position, trip purpose, or personal preferences not captured in the dataset.

The held-out prediction exercise also shows the limits of AI as a predictive model. On 50 held-out search queries, the AI agent correctly predicted 9 choices, for an exact-choice accuracy of 18%. The MNL model correctly predicted 11 choices, for an exact-choice accuracy of 22%. Under this strict metric, the MNL slightly outperformed the AI agent. This is not surprising because the MNL was estimated directly from observed human booking data, while the AI agent only received a small number of examples in context.

At the same time, exact-choice accuracy is a high bar in a discrete choice setting because many hotel alternatives are similar and customer choices are noisy. The behavioral comparison gives a more nuanced result. The observed purchase rate in the held-out data was 74%. The AI predicted a purchase rate of 82%, while the MNL hard-choice rule predicted a purchase rate of only 8%. This suggests that although the AI often missed the exact hotel chosen, its aggregate purchase/no-purchase behavior was closer to the observed data than the MNL hard-choice predictions. When the full MNL probability distribution was used instead of only the highest-probability hard choice, the MNL-implied purchase rate rose to about 65.6%, much closer to the observed behavior. This shows that probability-based summaries can be more informative than hard predictions alone.

Overall, the findings suggest that AI agents are useful for simulating plausible, directionally reasonable hotel choice behavior, especially when the goal is to generate choices that reflect common-sense preferences over observable attributes. However, AI agents are less reliable as exact predictors of individual human choices. Human booking behavior is noisy and depends on many factors not included in the prompt or dataset. Therefore, AI-generated choices should be viewed as a useful benchmark or behavioral simulation tool, but not as a replacement for real choice data or statistically estimated demand models. The strongest use case for AI in this setting may be as a complementary tool: it can generate plausible synthetic behavior and provide qualitative intuition, while MNL-style models provide disciplined estimation and probability-based predictions from observed data.
