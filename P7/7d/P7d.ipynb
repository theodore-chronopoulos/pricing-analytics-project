## 7d. Discussion

The results suggest that AI agents can capture some broad patterns in hotel choice behavior, but they are not a perfect substitute for models estimated from real human choices. In the AI-generated dataset, the choices implied intuitive preferences: the AI tended to favor hotels with higher star ratings, better review scores, better location scores, brand status, accessibility, and promotions, while penalizing higher displayed prices and higher historical prices. These signs are broadly consistent with the human MNL estimates, suggesting that the AI agent captures the basic economic tradeoffs in the hotel search setting.

However, the AI-generated choices appeared more deterministic and more strongly attribute-driven than real human choices. When the MNL was re-estimated on the AI-generated booking column, the AI-implied coefficients appeared stronger in magnitude, though the exact magnitudes are not directly comparable. Because the AI estimates were regularized and based on a 500-query sample, I interpret these results mainly by signs and qualitative patterns rather than exact coefficient magnitudes. The main takeaway is that the AI relied heavily on observable features such as price, review score, location, and star rating, whereas real customers likely respond to unobserved factors such as hotel photos, loyalty programs, brand familiarity, map position, trip purpose, or personal preferences not captured in the dataset.

The held-out prediction exercise shows the limits of AI as an exact predictive model. On 250 held-out search queries, the AI agent correctly predicted 33 choices, for an exact-choice accuracy of 13.2%. The MNL hard-choice rule correctly predicted 74 choices, for an accuracy of 29.6%. However, the always-`NO_PURCHASE` baseline also correctly predicted 74 choices, for the same accuracy of 29.6%. Therefore, the MNL hard-choice rule did not outperform the trivial no-purchase baseline on this held-out sample. Because exact-choice prediction is difficult when each query contains many similar hotel alternatives, these accuracy results should be interpreted cautiously.

At the same time, exact-choice accuracy is a high bar in a discrete choice setting because many hotel alternatives are similar and customer choices are noisy. The behavioral comparison gives a more nuanced result. The observed purchase rate in the held-out data was 70.4%. The AI predicted a purchase rate of 90.4%, while the MNL hard-choice rule predicted a purchase rate of only 0.8%. This suggests that although the AI often missed the exact hotel chosen and over-predicted purchases, its aggregate purchase/no-purchase behavior was still directionally closer to the observed data than the hard-choice MNL rule.

The MNL model becomes more informative when evaluated using its full probability distribution rather than only its highest-probability hard choice. The MNL probability-implied purchase rate was about 69.1%, much closer to the observed 70.4% than the hard-choice MNL purchase rate of 0.8%. Its expected chosen-hotel characteristics were also closer to observed behavior than the hard-choice MNL summary. This shows why probability-based summaries can be more informative than hard predictions alone in discrete choice settings.

Overall, the findings suggest that AI agents are useful for simulating plausible, directionally reasonable hotel choice behavior, especially when the goal is to generate choices that reflect common-sense preferences over observable attributes. However, AI agents are less reliable as exact predictors of individual human choices. Human booking behavior is noisy and depends on many factors not included in the prompt or dataset. Therefore, AI-generated choices should be viewed as a useful benchmark or behavioral simulation tool, but not as a replacement for real choice data or statistically estimated demand models. The strongest use case for AI in this setting may be as a complementary tool: it can generate plausible synthetic behavior and qualitative intuition, while MNL-style models provide disciplined estimation and probability-based predictions from observed data.
